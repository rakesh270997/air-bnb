```python
#import Pandas

import pandas as pd

# creating dataframe

df = pd.read_csv("E:/DATA/PYTHON DATASETS/Airbnb_Open_Data.csv")

#print the dataframe
df
```

```python
# now apply groupby() function
# group the data on review rate number

gk = df.groupby('review rate number')
```

```python
# selecting rows based on condition 
rslt_df = df.loc[df["review rate number"] == 5] 
    
print('\nResult dataframe :\n')

print(len(df), len(rslt_df))
```

```python
flt_no_rows = len(rslt_df)

totl_no_rows = len(df)

avg = ( flt_no_rows / totl_no_rows) * 100

print(int(avg), "%")
```

```python
#cost

rslt_df["price"] = rslt_df["price"].replace("[$,]", "", regex=True).astype(int)

sum_cost = rslt_df['price'].sum()

avg = ( sum_cost / flt_no_rows)
```
